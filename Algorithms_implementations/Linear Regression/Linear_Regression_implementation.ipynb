```python
import numpy as np 
class LinearRegression : 
    def __init__(self , learning_rate = 0.01 , epochs = 1000) : 
        self.learning_rate = learning_rate
        self.epochs = epochs 
        self.m = 0 # initializing the scope 
        self.b = 0 # initializing the intercept 
    def fit(self , X , Y) : 
        n = len(X)
        for _ in range(self.epochs) : 
            Y_predict = self.m*X + self.b
            error = Y_predict - Y 
            dm = (2/n) * np.sum(error*X)
            db = (2/n) * np.sum(error)
            self.m -= dm*self.learning_rate
            self.b -= db*self.learning_rate
    def predict(self , X ): 
        return X * self.m + self.b
```

```python
import numpy as np

class LinearRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations

    def fit(self, X, Y):
        self.m, self.n = X.shape
        self.W = np.zeros(self.n)
        self.b = 0
        self.X = X
        self.Y = Y

        for _ in range(self.iterations):
            self.updateweights()

        return self

    def updateweights(self):
        Y_pred = self.predict(self.X)

        dW = - (2 * self.X.T.dot(self.Y - Y_pred)) / self.m
        db = - (2 * np.sum(self.Y - Y_pred)) / self.m

        self.W -= self.learning_rate * dW
        self.b -= self.learning_rate * db

    def predict(self, X):
        return X.dot(self.W) + self.b
```

```python
# Generate sample data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
Y = 4 + 3 * X + np.random.randn(100, 1)  # Y = 4 + 3X + noise

# Train the model
model = LinearRegression()
model.fit(X, Y)

print(f"Final slope (m): {model.m}")
print(f"Final intercept (b): {model.b}") 

# Output:
# Final slope (m): 2.799408338101361
# Final intercept (b): 4.181917499468203
```
